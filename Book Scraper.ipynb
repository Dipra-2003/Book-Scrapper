# Scrapping from book

```python
#For accessing all the pages

data=[]
for i in tqdm(range(1,51)):
    multilink='https://books.toscrape.com/catalogue/page-' + str(i) +'.html'
    multires=requests.get(multilink)
    multisoup=BeautifulSoup(multires.text,'html.parser')
    
    for sp in multisoup.find_all('li',class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
   
    #print(sp.find_all('a')) #Finding anchor tags
    
    #print(sp.find_all('a')[-1]) #Finding only the last tag for further use
    
    #Printing the link and title of the books
        book_link='https://books.toscrape.com/catalogue/'+ sp.find_all('a')[-1].get('href') 
        book_title=sp.find_all('a')[-1].get('title')
        book_image='https://books.toscrape.com' + sp.find('img').get('src')[2:]
        book_rating=sp.find('p').get('class')[-1]
        book_cost=sp.find('p',class_='price_color').text[1:]
        book_avail=sp.find('p',class_='instock availability').text.strip()
    
        
        #print("Book Title-",book_title,"\nBook Link-",book_link,"\nBook Image-",book_image,"\nBook Rating-",book_rating,"\nBook Cost-",book_cost,"\nAvailability-",book_avail)
        data.append([book_title,book_rating,book_cost,book_avail,book_image,book_link])
        
        
```

```python
df=pd.DataFrame(data,columns=['Book_Title','Book_Rating','Book_Cost','Book_Avail','Book_Image','Book_Link'])
df
```

```python
#To save in folder
df.to_csv('Book-Scrapper.csv',index=False)
```

# Scrapping from pages

```python
#Scrapping from the pages
page=[]

for link in tqdm((df['Book_Link'])):
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')
    book_type= soup.find('ul',class_='breadcrumb').find_all('a')[-1].text
    UPC=soup.find('table',class_='table table-striped').find_all('td')[0].text
    Price_exc_Tax=soup.find('table',class_='table table-striped').find_all('td')[2].text[2:]
    Price_inc_Tax=soup.find('table',class_='table table-striped').find_all('td')[3].text[2:]
    Tax=soup.find('table',class_='table table-striped').find_all('td')[4].text[2:]
    Availability=soup.find('table',class_='table table-striped').find_all('td')[5].text
    Numberof_reviews= soup.find('table',class_='table table-striped').find_all('td')[6].text

    
    page.append([UPC,Price_exc_Tax,Price_inc_Tax,Tax,Availability,Numberof_reviews])

    #break
```

```python
pages=pd.DataFrame(page,columns=['UPC','Price_exc_Tax','Price_inc_Tax','Tax','Availability','Numberof_reviews'])
pages
```

```python
pages.to_csv('Page-Scrapper.csv',index=False)
```
